```python
import plotly
import matplotlib.pyplot as plt
import numpy as np
```

```python
gapminder_data = plotly.data.gapminder()
pop_year_df = gapminder_data[['pop', 'year']]
```

```python
fig = pop_year_df.plot.line(x='year', y='pop').figure
plt.title("Population Trends Over Time")
plt.xlabel("Years")
plt.ylabel("Population")
fig.savefig('to.png')
plt.show()
plt.close(fig)
```
