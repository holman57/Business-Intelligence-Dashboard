# Accessing gapminder datasets from `plotly` import

```python
import plotly
import matplotlib.pyplot as plt
import numpy as np
```

```python
gapminder_data = plotly.data.gapminder()
```

```python
fig = gapminder_data['continent'].value_counts().plot(kind='bar').figure
print(gapminder_data['continent'].value_counts())
plt.xlabel("Continents")
plt.xticks(rotation=30, horizontalalignment="center")
plt.title("Continents in the Gapminder dataset")
plt.ylabel("Number of instances")
fig.savefig('bars.png')
plt.show()
plt.close(fig)
```

```python
pop_year_df = gapminder_data[['pop', 'year']]
fig = pop_year_df.plot.line(x='year', y='pop').figure
plt.title("Population Trends Over Time")
plt.xlabel("Years")
plt.ylabel("Population")
fig.savefig('lines.png')
plt.show()
plt.close(fig)
```

```python
us_gapminder_df = gapminder_data.loc[gapminder_data['country'] == "United States"]
pop_year_df = us_gapminder_df[['pop', 'year']].sort_values(by='year')
fig = pop_year_df.plot.line(x='year', y='pop').figure
plt.title("Population Trends Over Time for the US")
plt.xlabel("Years")
plt.ylabel("Population")
fig.savefig('line.png')
plt.show()
plt.close(fig)
```

```python
fig = us_gapminder_df.plot.scatter(x='lifeExp', y='pop', c='gdpPercap', colormap='viridis').figure
plt.xlabel("Life Expectancy")
plt.title("Life Expectancy and Population for the US")
plt.ylabel("Population")
fig.savefig('scatter.png')
plt.show()
plt.close(fig)
```

```python
fig = gapminder_data.plot.scatter(x='lifeExp', y='pop', c='gdpPercap', colormap='viridis').figure
plt.xlabel("Life Expectancy")
plt.title("Life Expectancy and Population")
plt.ylabel("Population")
fig.savefig('cluster_scatter.png')
plt.show()
plt.close(fig)
```

```python
fig,ax = plt.subplots()

for name in ['United States','India', 'Austria']:
    #Create subset
    df = gapminder_data.loc[gapminder_data['country'] == name]
    #Create dataframe with population and year
    pop_year_df = df[['pop', 'year']].sort_values(by='year')
    #Plot a line plot
    ax.plot(pop_year_df['year'], pop_year_df['pop'], label=name)

ax.set_xlabel("Year")
ax.set_ylabel("Population")
plt.title("Population Over Time")
ax.legend(loc='best', title="Country")
fig.savefig('comparison.png')
plt.show()
plt.close(fig)
```

```python
fig,ax = plt.subplots()

for name in ['United States','India', 'Austria']:
    df = gapminder_data.loc[gapminder_data['country'] == name]
    #Generate scatter plot
    ax.scatter(df['gdpPercap'], df['pop'], label=name)

ax.set_xlabel("gdpPercap")
ax.set_ylabel("Population")
ax.legend(loc='best', title="Country")
fig.savefig('scatter_comparison.png')
plt.show()
plt.close(fig)
```

```python
fig,ax = plt.subplots()

for name in ['United States','India', 'Austria']:
    df = gapminder_data.loc[gapminder_data['country'] == name]
    ax.scatter(df['lifeExp'], df['pop'], label=name)

ax.set_xlabel("lifeExp")
ax.set_ylabel("Population")
ax.legend(loc='best', title="Country")
fig.savefig('cluster_scatter_comparison.png')
plt.show()
plt.close(fig)
```

```python
gapminder_data = plotly.data.gapminder()
#Get the row of the minimum population value, across all years in the data
minimum_pop = gapminder_data.loc[gapminder_data['pop'] == gapminder_data['pop'].min()]
print(minimum_pop)
#Get the row of the maximum population value, across all years in the data
max_pop = gapminder_data.loc[gapminder_data['pop'] == gapminder_data['pop'].max()]
print(max_pop)
```

We can see here that the country with the lowest population in the year 1952 was Sao Tome and Principe, and China was the country with the highest population in the year 2007.

```python
fig,ax = plt.subplots()

for name in ['India','China', 'Sao Tome and Principe']:
    df = gapminder_data.loc[gapminder_data['country'] == name]
    #Generate scatterplot
    ax.scatter(df['lifeExp'], df['pop'], label=name)

ax.set_xlabel("lifeExp")
ax.set_ylabel("Population")
ax.legend(loc='best', title="Country")
fig.savefig('correlations.png')
plt.show()
plt.close(fig)
```
